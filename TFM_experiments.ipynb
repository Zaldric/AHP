```python
DRIVE_PATH = '/content/gdrive/MyDrive/TFM/datasets'

TOKENIZER = 'distilbert-base-uncased'

MODEL_PATH = f'/content/gdrive/MyDrive/TFM/models'
SAMPLE_SIZE = 25000
```

```python
from google.colab import drive


drive.mount('/content/gdrive')
```

```python
from transformers import AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained(TOKENIZER)
```

```python
import torch
from transformers import AutoModel


twitter_model = AutoModel.from_pretrained(f"{MODEL_PATH}/Twitter/Twitter-{SAMPLE_SIZE}")
amazon_model = AutoModel.from_pretrained(f"{MODEL_PATH}/Amazon/Amazon-{SAMPLE_SIZE}")
imdb_model = AutoModel.from_pretrained(f"{MODEL_PATH}/IMDB/IMDB-{SAMPLE_SIZE}")

device = torch.device("cuda:0")
```

```python
# Run inferences with your new model using Pipeline
from transformers import pipeline


twitter_pipeline = pipeline('sentiment-analysis', model=twitter_model, tokenizer=tokenizer, device=device)
amazon_pipeline = pipeline('sentiment-analysis', model=amazon_model, tokenizer=tokenizer, device=device)
imdb_pipeline = pipeline('sentiment-analysis', model=imdb_model, tokenizer=tokenizer, device=device)

twitter_pipeline("I hate my life")
amazon_pipeline("I hate my life")
imdb_pipeline("I hate my life")
```
